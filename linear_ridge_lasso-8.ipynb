```python
# importing necessary libraries
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_absolute_error
```

```python
# load the data
df = pd.read_csv('House_price_prediction.csv')
df.head()
```

```python
# dropping first column (Unnamed: 0)
df = df.drop('Unnamed: 0', axis=1) # axis=0 for row, axis=1 for column
df.head()
```

**One-Hot Encoding**: is a technique used to convert categorical (text) data into numerical data so that machine learning algorithms can understand it.

```python
# One-Hot Encoding:
df = pd.get_dummies(df, drop_first=True, dtype=int)
df.head()
```

```python
# dividing dataframe into X (features) and y (target)
X = df.drop('Price', axis=1)
y = df['Price']

# splitting into training & testing data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=29)
```

```python
# Linear Regression
LR_model = LinearRegression()
LR_model.fit(X_train, y_train)
```

```python
# to check efficiency
print(f'Training efficiency score: {round(LR_model.score(X_train, y_train) * 100, 2)}')
print(f'Testing efficiency score: {round(LR_model.score(X_test, y_test) * 100, 2)}')
# Overfitting
```
