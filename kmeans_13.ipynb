**K-Means** Algorithm: **Unsupervised** Learning

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans # KMeans Algorithm from the scikit-learn library
from sklearn.datasets import load_iris
```

```python
data = load_iris()
X = data.data
```

**Inertia** (also called **Within-Cluster Sum of Squares(WCSS)**) is a measure of how compact the clusters are.<br>
It is the sum of the squared distances between each data point and the centroid of its assigned clusters.

```python
# Elbow method: used to find optimized k value

# i) find inertia values
out = []
for i in range(1, 11): # 10 different k-means model
  model = KMeans(n_clusters=i)
  model.fit(X)
  out.append(model.inertia_) # model.inertia_ stores the Inertia
# display(out)

# ii) locating k value
from kneed import KneeLocator
# optimal number of clusters (K) using the KneeLocator class from the kneed library
k_values = range(1, 11) # K from 1 to 10
KL = KneeLocator(k_values, out, curve='convex', direction='decreasing')
k = KL.elbow
print("Optimized k-value: ", k)

# iii) plotting
plt.style.use('dark_background')
plt.plot(k_values, out, color='red', marker='o', mfc='blue', mec='white')
plt.xlabel("K-values")
plt.ylabel("Inertia")
plt.title("Elbow Curve", color='green')
plt.show()
```

```python
# kmeans model
model = KMeans(n_clusters=k)
model.fit(X)
```
