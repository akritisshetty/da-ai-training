**DATA CLEANING FOR HR DATASET**

```python
# Step 0: Import the dataset
df = pd.read_csv('hr_data.csv')
df.head(2)
```

```python
# Step 1: Renaming the columns
df.rename(columns={
    'employee_id' : 'eid',
    'number_project' : 'no._of_projects',
    'average_montly_hours' : 'avg_mon_hours',
    'time_spend_company' : 'time_spent_cpy',
    'Work_accident' : 'work_accident',
    'left' : 'left_company',
    'promotion_last_5years' : 'promo_last_5yrs',
    'department' : 'dept',
    'salary' : 'sal'
}, inplace=True)
df.head(2)
```

```python
# Step 2: Rearranging the columns
new_order = [0, 7, 1, 2, 3, 6, 8, 4, 5]
df[df.columns[new_order]].head(2)
```

```python
# Step 3: Updating the values in the columns
df['dept'].unique()

dict_var = {
    'sales' : 'Sales D',
    'accounting' : 'Accounts D',
    'hr' : 'HR D',
    'technical' : 'Tech D',
    'support' : 'TechSupport D',
    'management' : 'Mgmt D',
    'IT' : 'IT D',
    'product_mng' : 'ProMgmt D',
    'marketing' : 'Market D',
    'RandD' : 'RandD D'
}
df['dept'] = df['dept'].map(dict_var)
df.head(80)
```

```python
# Step 4: Removing unwanted data from the column
df['dept'] = df['dept'].str.strip(' D')
df
```

```python
# 5. Changing the datatype of the column
# display(df.info())
df['avg_mon_hours'] = df['avg_mon_hours'].astype(float)
df.info()
```

```python
# 6. Changing particular value in column
df['dept'] = df['dept'].str.replace('TechSupport', 'Support')
df.tail(3)
```

```python
# 7. Applying in-built functions on column
df['sal'] = df['sal'].str.upper()
df.head(3)
```

```python
# 8. Handling missing values
# i) dropping NaN values
# ii) replacing NaN values with 0 or avg
# iii) forward filling or back filling
# iv) dropping duplicates
df.dropna()
df.fillna(0)
df.fillna('ffill')
df.fillna('bfill')
df.drop_duplicates(inplace = True)
df
```

```python
# 9. Removing unwanted column from the dataframe
df.drop('dept', axis=1)
df.head(3)
```

```python
# 10. Removing unwanted rows
df.drop(1, axis=0, inplace=True)
df.head(3)
```

```python
# 11. Resetting index
df.reset_index() # resets the index and displays old index in 'index' column in the dataframe
df.head()
```

```python
# 12. Setting column name as index
df.set_index('eid')
```
