```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

**matplotlib Limitations:**<br>
1. no advanced colour customization<br>
2. more lines of code<br>
3. cannot perform advanced visualization<br>

**seaborn Advantages:**<br>
1. creates plots with less code<br>
2. advanced colour customization<br>
3. advanced visualizations<br>
4. supports variety of charts


```python
df = pd.read_csv('winequalityN.csv')
df.head(3)
```

```python
df['quality'] = df['quality'].map({3:'Bad', 4:'Bad', 5:'Average', 6:'Good', 7:'Good', 8:'Good', 9:'Excellent'})
```

**Histplot**

```python
plt.style.use('dark_background')
sns.histplot(x='fixed acidity', data=df, kde=True) # kernal density estimation: x--numeric, y--count by default
plt.title('Distribution of Fixed Acidity')
plt.xlabel('Fixed Acidity')
plt.ylabel('Frequency')
plt.show()
```

**Boxplot** : to identify the outliers

```python
sns.boxplot(x='type', y='pH', data=df, hue='type', palette='inferno', flierprops={'mfc':'red', 'mec':'white'})
plt.show()
# palette: name of the colour map
# flierprops: divide the categorical column
```

**Violin Plot** : Used to find where exactly more number of data points are present.

```python
sns.violinplot(x='type', y='pH', data=df, hue='type', palette='inferno')
plt.show()
```

**Count Plot** : Used to find the count based on categorical column.

```python
a = sns.countplot(x='type', data=df, hue='type', palette='inferno', width=0.5)
for i in a.containers: # in-built attribute which holds multiple containers
  a.bar_label(i)
plt.title('Count by Wine Type')
plt.show()
# x is categorical column, y is counted by default
```

**Pair Plot** : to understand the distribution of data based on multiple columns.

```python
data = df.iloc[:, 0:5] # iloc is used for slicing: all rows, 0 to 4 column
sns.pairplot(data, hue='type', palette='inferno')
plt.show()
```
