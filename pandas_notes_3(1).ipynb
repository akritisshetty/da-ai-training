**Boolean Filtering**

```python
df = pd.read_csv('titanic.csv')
df.head()
```

```python
print("Filter only survived from survived column:")
df['survived'] == 1
```

```python
print("Survived and male:")
df[(df['survived'] == 1) & (df['sex'] == 'male')]
```

```python
print("Get female, survived who are in pclass = 3")
df[(df['sex'] == 'female') & (df['survived'] == 1) & (df['pclass'] == 3)]
```

```python
print("Get passenges details who are either female or not survived")
df[(df['sex'] == 'female') | (df['survived'] == 0)]
```

```python
print("Get details of passengers embarked in either S or C")
df[(df['embarked'] == 'S') | (df['embarked'] == 'C')]
```

**DATA CLEANING**

```python
data = pd.read_csv("amazon_fires_.csv", encoding = 'ISO-8859-1') # encoding changes the Spanish in file to English for Colab to understand
data
```

```python
# 2. Rearranging the columns
new_order = [2, 1, 0, 4, 3]
df[df.columns[new_order]]
```

```python
# 1. Renaming the columns
data.rename(columns={
    'ano': 'year',
    'mes': 'month',
    'estado': 'state',
    'numero': 'no. of fires',
    'encontro': 'meeting date'
}, inplace=True)

data
```

```python
# 3. Updating values inside the column
dict_var = {
    'Janeiro': 'January',
    'Fevereiro': 'February',
    'Março': 'March',
    'Abril': 'April',
    'Maio': 'May',
    'Junho': 'June',
    'Julho': 'July',
    'Agosto': 'August',
    'Setembro': 'September',
    'Outubro': 'October',
    'Novembro': 'November',
    'Dezembro': 'December'
}

data['month'] = data['month'].map(dict_var)

data
```

```python
# 4. Removing unwanted data from the column
data['no. of fires'] = data['no. of fires'].str.strip(' Fires')
data
```

```python
# 5. Changing the datatype of the column
data['no. of fires'] = data['no. of fires'].astype(float)
data
```

```python
# 6. Changing particular value in column
data['month'] = data['month'].str.replace('January', 'Jan')
data['month'] = data['month'].str.replace('December', 'Dec')
data
```

```python
# 7. Applying in-built functions on column
data['state'] = data['state'].str.upper()
data.head(3)
```

```python
# 8. Handling missing values
# i) dropping NaN values
# ii) replacing NaN values with 0 or avg
# iii) forward filling or back filling
# iv) dropping duplicates
data.dropna()
data.fillna(0)
data.fillna('ffill')
data.fillna('bfill')
data.drop_duplicates(inplace = True)
data
```

**concat() & merge()**

```python
# concatenating two dataframes
df1 = pd.DataFrame({'ename': ['A', 'B', 'C'], 'dept': ['CSBS', 'ECE', 'AIML']})
df2 = pd.DataFrame({'dept': ['AIML', 'CSD'], 'marks': [98, 96]})
```

```python
pd.concat([df1, df2], ignore_index=True) # row-wise
```

```python
pd.concat([df1, df2], ignore_index=True, axis=1) # column-wise
```

```python
# merging two dataframes
df1 = pd.DataFrame({'ename': ['A', 'B', 'C'], 'dept': ['CSBS', 'ECE', 'AIML']})
df2 = pd.DataFrame({'dept': ['AIML', 'CSD'], 'marks': [98, 96]})

# merge - same as inner join
pd.merge(df1, df2, on='dept')
```

```python
# inner join
pd.merge(df1, df2, on='dept', how='inner')
```

```python
# outer join
pd.merge(df1, df2, on='dept', how='outer')
```

```python
# left join
pd.merge(df1, df2, on='dept', how='left')
```

```python
# right join
pd.merge(df1, df2, on='dept', how='right')
```
