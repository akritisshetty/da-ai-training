```python
import pandas as pd
import numpy as np
```

**Linear Regression Model & Multiple Linear Regression**

```python
# sklearn: scikitlearn: ML Library
from sklearn.linear_model import LinearRegression
# LinearRegression is a class, linear_model is the module (each library has multiple modules, modules have classes, classes have functions)
from sklearn.model_selection import train_test_split # train_test_split is a function
from sklearn.datasets import fetch_california_housing # dataset
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
# r2score = measures how well the regression model fits the data
# mean_squared_error = calculates average of the squared differences between actual and predicted values
# mean_absolute_error = measures the average absolute difference between actual and predicted values
# Lower MAE indicates better predictions
```

A linear model assumes there is a linear relationship between the input feature (x) and the target (y).

```python
# importing data
data = fetch_california_housing()
X = data.data
y = data.target
df = pd.DataFrame(X, columns=data.feature_names)
df['Target'] = y
df.head()

# splitting data
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size=0.2)
```

```python
# creating model object of class LinearRegression
model = LinearRegression()
model.fit(Xtrain, ytrain)
```

```python
# to observe intercept & slope (in-built parameter)
display(model.intercept_)
display(model.coef_)
```

```python
# prediction
ypred = model.predict(Xtest)

# checking model performance
print("Performance Metrics")
print('R2 Score: ', round(r2_score(ytest, ypred), 4))
print('MSE: ', round(mean_squared_error(ytest, ypred), 4))
print('MAE: ', round(mean_absolute_error(ytest, ypred), 4))
```
