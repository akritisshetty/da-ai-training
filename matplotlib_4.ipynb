```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# DataFrame for Website Visitors
df = pd.DataFrame({
    'Day': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'],
    'Total_Visitors': [120, 150, 170, 140, 180, 220, 200],
    'Mobile_Visitors': [80, 100, 120, 110, 140, 180, 160]

})
df
```

```python
plt.style.use('dark_background') # to get black background
```

**Line Chart** : Used to visualize how the values are changing over time

```python
# How visitors are changing over time
# BASIC (linestyle - solid)
plt.plot(df['Day'], df['Total_Visitors'])
plt.show()
```

```python
#Added color, marker
plt.plot(df['Day'], df['Total_Visitors'], color='green', marker='o')
plt.show()
```

```python
# different marker
plt.plot(df['Day'], df['Total_Visitors'], color='green', mfc='red', mec='white', marker='p') # pentagonal markers
plt.show()
# mfc: marker face color
# mec: marker edge color
```

```python
# square marker
plt.plot(df['Day'], df['Total_Visitors'], color='green', mfc='red', mec='white', marker='s') # square markers
plt.show()
```

```python
# different linestyle - dotted
plt.plot(df['Day'], df['Total_Visitors'], linestyle='dotted', color='green', marker='o', markersize=8, linewidth=3) # circle markers, markersize, linewidth, linestyle
plt.show()
```

```python
# different linestyle - dashed
plt.plot(df['Day'], df['Total_Visitors'], linestyle='dashed', color='green', marker='o', markersize=8, linewidth=3) # circle markers, markersize, linewidth, linestyle
plt.show()
```

```python
# linestyle: deafault(-), '--', '-', '-.', ':', 'solid', 'dashed', 'dashdot', 'dotted'
# markers: 'o', '+', '*', 'x', '.', '_', '|', '^', 'v', '<', '>', 'square', 's', 'diamond', 'd', 'pentagon', 'p', 'hexagon', 'h'
```

```python
# Plotting 2 lines together
plt.plot(df['Day'], df['Total_Visitors'], linestyle='dashdot', linewidth=3, color='green', mfc='red', mec='white', marker='s', label='Total_Visitors')
plt.plot(df['Day'], df['Mobile_Visitors'], color='red', mfc='green', mec='white', marker='d', label='Mobile_Visitors')
plt.legend() # storing & displaying the labels
plt.show()
```

```python
# Plotting 2 lines together
plt.plot(df['Day'], df['Total_Visitors'], linestyle='dashdot', linewidth=3, color='green', mfc='red', mec='white', marker='s', label='Total_Visitors')
plt.plot(df['Day'], df['Mobile_Visitors'], color='red', mfc='green', mec='white', marker='d', label='Mobile_Visitors')
plt.xlabel('Day')
plt.ylabel('Number of Visitors')
plt.title('Website Visitors')
plt.legend() # storing & displaying the labels
plt.show()
```

**Bar Chart** : Used to compare values across different categories. <br>
plt.figure(figsize = (width,height))

```python
# DataFrame for Products & Revenue
df = pd.DataFrame({
    'product': ['laptop', 'mobile', 'tablet', 'headphones', 'smartwatch', 'camera', 'printer', 'monitor', 'keyboard', 'mouse'],
    'units_sold': [45, 120, 60, 85, 70, 30, 25, 50, 90, 110],
    'revenue': [2475000, 2400000, 1522222, 255000, 350000, 1350000, 300000, 750000, 135000, 88000, ]
})
df
```

```python
# 1. Which product sold more?
# 2. Which product sold less?
plt.figure(figsize=(12,5))
fig = plt.bar(df['product'], df['units_sold'], color='lightgreen', width=0.5)
plt.bar_label(fig) # to get bar labels (45, 120, 60, 85, 70, 39, 25, 50, 90, 110)
plt.xlabel('Products')
plt.ylabel('No. of Units')
plt.title('Sales by Products')
plt.show()
```

```python
# Compare the revenue
plt.figure(figsize=(12,5))
fig = plt.bar(df['product'], df['revenue'], color='pink', width=0.5)
plt.bar_label(fig) # to get bar labels
plt.xlabel('Products')
plt.ylabel('Revenue')
plt.title('Sales by Revenue')
plt.show()
```

**Pie Chart** : Used to visualize the percentage. Can display only one series of data.

```python
# DataFrame
df = pd.DataFrame({
    'order_id': [101, 102, 103, 104, 105, 106, 107, 108],
    'item': ['pizza', 'borgir', 'pasta', 'pizza', 'borgir', 'pizza', 'pasta', 'borgir'],
    'category': ['fastfood', 'fastfood', 'italian', 'fastfood', 'fastfood', 'fastfood', 'italian', 'fastfood'],
    'quantity': [2, 1, 1, 3, 2, 1, 2, 1],
    'revenue': [800, 120, 280, 1200, 240, 400, 560, 120]
})
df
```

```python
# Sum of revenue per item
data = df.groupby('item')['revenue'].sum()
data
```

```python
# What percentage of total revenue comes from each food item?
plt.pie(data, labels=data.index, colors=['red', 'pink', 'lightgreen'], explode=(0, 0, 0), autopct='%0.1f%%')
# explode: separates each section
# autopct: used to tell how many floating point digits should be displayed on the pie chart (considers only 1 floating point value)
plt.show()
```

```python
# Which category has largest proportion of revenue?

# Groupby revenue per category
data = df.groupby('category')['revenue'].sum()
data

# Plot pie chart
plt.pie(data, labels=data.index, colors=['forestgreen', 'maroon'], autopct='%0.1f')
plt.title("Revenue by Category")
plt.show()
```
